```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL halaman Berita Crypto
url = 'https://cryptoharian.com/7-alasan-kenapa-bull-bitcoin-belum-selesai-menurut-trader/'

# Mengirim permintaan GET ke URL
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

# Menemukan elemen HTML yang mengandung ulasan
reviews = soup.find_all('div', class_='box-inner-p-bigger box-single')  # Sesuaikan dengan kelas yang sebenarnya

# Menyiapkan list untuk menyimpan data ulasan
data = []

for review in reviews:
    judul = review.find('h1', class_='entry-title').text  # Sesuaikan dengan kelas yang sebenarnya
    berita = review.find('div', class_='entry-content') # Sesuaikan dengan kelas yang sebenarnya
    print(f'Judul: {judul}')
    print(f'Berita: {berita.text}')
    
    # Periksa apakah elemen ditemukan sebelum mengakses atribut text
    if judul is not None and berita is not None:
        data.append({
            'Judul': judul,
            'Berita': berita.text
        })

# Membuat DataFrame dan menyimpannya ke dalam CSV
df = pd.DataFrame(data)
df.to_csv('cryptoharian.csv', index=False)
df = pd.read_csv('./Crypto.csv')
df
```
