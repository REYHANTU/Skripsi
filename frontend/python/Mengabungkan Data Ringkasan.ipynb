```python
import pandas as pd
import glob

# Daftar path file CSV yang ingin digabungkan
file_paths = glob.glob('preprocessed/*.csv')  # Ganti dengan path folder berisi file CSV kamu

# Membaca dan menggabungkan semua file CSV
dataframes = []
for file_path in file_paths:
    df = pd.read_csv(file_path)
    dataframes.append(df)

# Menggabungkan semua dataframe
combined_df = pd.concat(dataframes, ignore_index=True)

# Menyimpan hasil gabungan ke file CSV
combined_df.to_csv('./preprocessed/merge_datasets.csv', index=False)

print("Penggabungan selesai. Dataset telah disimpan sebagai 'merge_datasets.csv'.")

# Mencetak hasil gabungan file CSV
df = pd.read_csv('./preprocessed/merge_datasets.csv')
print(df)
```

```python
import pandas as pd
import glob

# Daftar path file CSV yang ingin digabungkan
file_paths = glob.glob('Datasets/*.csv')  # Ganti dengan path folder berisi file CSV kamu

# Membaca dan menggabungkan semua file CSV
dataframes = []
for file_path in file_paths:
    df = pd.read_csv(file_path)
    dataframes.append(df)

# Menggabungkan semua dataframe
combined_df = pd.concat(dataframes, ignore_index=True)

# Menyimpan hasil gabungan ke file CSV
combined_df.to_csv('./Datasets/combined_datasets.csv', index=False)

print("Penggabungan selesai. Dataset telah disimpan sebagai 'combined_datasets.csv'.")

# Mencetak hasil gabungan file CSV
df = pd.read_csv('./Datasets/combined_datasets.csv')
print(df)
```

```python
import os
import pandas as pd

# Ensure the directory exists
output_dir = './preprocessed/'
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Load the CSV file
df = pd.read_csv('./preprocessed/merge_datasets.csv')

# Clean up the third column header by removing extra semi-colons
df.columns = ['Title', 'Date', 'Summary']

# Clean the 'Summary' column by removing any extra semi-colons and trimming whitespace
df['Summary'] = df['Summary'].str.replace(';;;', '', regex=False).str.strip()

# Save the cleaned DataFrame to a new CSV file
df.to_csv(os.path.join(output_dir, 'cleaned_merge_datasets.csv'), index=False)

# Show the cleaned dataframe
df.head()
```
